```python
# ================================
# ATTENDANCE MANAGEMENT SYSTEM
# ================================

# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


students = np.array(["PRAKHAR", "PIKENDRA", "KRISH", "AMAN", "VAIBHAV"])



attendance = np.array([
    [1, 0, 0, 1, 0],   # Day 1
    [1, 0, 1, 1, 1],   # Day 2
    [1, 1, 1, 0, 1],   # Day 3
    [0, 0, 0, 1, 1],   # Day 4
    [1, 1, 1, 1, 0]    # Day 5
])


df = pd.DataFrame(
    attendance,
    columns=students,
    index=["Day1", "Day2", "Day3", "Day4", "Day5"]
)

print("\nAttendance Sheet:\n")
print(df)


# Step 4: Total Attendance
total_present = df.sum()
print("\nTotal Present Days:\n")
print(total_present)


# Step 5: Attendance Percentage

attendance_percentage = (total_present / len(df)) * 100
print("\nAttendance Percentage:\n")
print(attendance_percentage)


plt.figure()
plt.bar(attendance_percentage.index, attendance_percentage.values)
plt.xlabel("Students")
plt.ylabel("Attendance Percentage")
plt.title("Student Attendance Percentage")
plt.show()


daily_attendance = df.sum(axis=1)

plt.figure()
plt.plot(daily_attendance.index, daily_attendance.values, marker='o')
plt.xlabel("Days")
plt.ylabel("Number of Students Present")
plt.title("Daily Attendance Trend")
plt.show()


plt.figure()
for student in students:
    plt.plot(df.index, df[student], marker='o', label=student)
plt.xlabel("Days")
plt.ylabel("Attendance (1=Present, 0=Absent)")
plt.title("Individual Student Attendance Pattern")
plt.legend()
plt.show()
```
